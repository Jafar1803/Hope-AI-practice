```python
BMI=float(input("Enter the BMI Index:"))
if BMI<18.5:
    print("Underweight")
elif 18.5<=BMI<=24.9:
    print("Normal")
elif 25.0<=BMI<=29.9:
    print("Overweight")
else:
    print("Very Overweight")
```
