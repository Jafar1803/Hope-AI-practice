```python
# print Odd Numbers in the list
my_tup=(20,10,16,19,25,1,276,188)
for each_number in my_tup:
    if each_number%2!=0:
        print(f"{each_number} is odd")
```

```python
# print Even numbers in the list
my_tup=(20,10,16,19,25,1,276,188)
for each_number in my_tup:
    if each_number%2==0:
        print(f"{each_number} is even")
```
